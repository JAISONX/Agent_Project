Basic LLM Call - Building the core brain

```python
from openai import OpenAI
import os

API_KEY = os.getenv("OPENAI_API_KEY")
client = OpenAI(api_key=API_KEY)

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "user", "content": "calculate 5*10"}
    ]
)
print(response.choices[0].message.content)
```

Adding Decision Logic (Making an agent)

```python
def agent(user_input):
    if "weather" in user_input:
        return "The weather is sunny."
    elif "math" in user_input:
        return "Use Calculator tool"
    else:
        return "ask LLM"
    
print(agent("what is 5+5?"))
```

Adding Tool to the agent

```python
def calculate(expression):
    return eval(expression)

def agent(user_input):
    if "calculate" in user_input:
        expr = user_input.replace("calculate", "")
        return calculate(expr)
    else:
        return "ask LLM"

print(agent("calculate 5*7"))
```

Combine LLM + Tool

```python
def agent(user_input):
    if "calculate" in user_input:
        expr = user_input.replace("calculate", "")
        return calculate(expr)
    
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "user", "content": user_input}
        ]
    )
    return response.choices[0].message.content

user_input = input("Enter : ")
agent_response = agent(user_input)
print(agent_response)
```
